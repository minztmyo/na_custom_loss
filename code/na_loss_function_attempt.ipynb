```python
y = np.array(df['group'])
```

```python
y
```

```python
encoder = LabelEncoder()
encoder.fit(y)
encoded_Y = encoder.transform(y)
dummy_y = np_utils.to_categorical(encoded_Y)
```

```python
X = df.drop(['group'], axis = 1)
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X.shape
```

```python
dummy_y.shape
```

```python
x_train , x_test , y_train , y_test =train_test_split(X,
                            dummy_y,test_size =0.2, random_state=105)
```

```python
model = Sequential()
#model.add(tf.keras.layers.Masking(mask_value=0., input_dim = 199))
model.add(Dense(units=128, activation='relu', input_dim = 199)) # add layers to neural network
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=4, activation='softmax'))
model.compile(loss="categorical_crossentropy",
              optimizer= "adam")
```

```python
model.fit(x=x_train,
        y=y_train,
        epochs=50,
        validation_split = 0.2
        )
```

```python
def na_custom_loss(y_true, y_pred):
    y_true_mask = tf.where(tf.math.is_nan(y_true), tf.zeros_like(y_true), y_true)
    return K.categorical_crossentropy(y_pred, y_true_mask)
```
